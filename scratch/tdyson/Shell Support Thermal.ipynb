```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_A(P, l, k, Ti, Tf):
    return P*l / (k * (Tf**2 - Ti**2))

def get_P(A, l, k, Ti, Tf):
    return A / l * k * (Tf**2 - Ti**2)

def Rs(A, t):
    r = (A - t**2) / (2*t)
    R = (A + t**2) / (2*t)
    return r, R

def m2_to_cm2(m2):
    return m2 * 1e4
```

# 4 K to Still

```python
# 4 K to still
P = 0.001 # W
l = 0.177 # m
k = 0.2/4 # W/K^2m (NIST linear extrapolation to 0)
Ti = 0.7 # K
Tf = 4 # K
A = 2 / 1e4 # m^2
```

```python
print(get_P(A, l, k, Ti, Tf))
```

```python
A
```

```python
Rs(m2_to_cm2(A(P, l, k, Ti, Tf))/4, 0.2)
```

```python
Ps = np.linspace(0, 0.01)
As = A(Ps, l, k, Ti, Tf)/4
```

```python
plt.plot(Ps, m2_to_cm2(As))
plt.grid('on')
```

```python
ts = np.linspace(0.1, 1)
PP, tt = np.meshgrid(Ps, ts)
Rs_grid = Rs(m2_to_cm2(A(PP, l, k, Ti, Tf)/4), tt)
```

```python
plt.imshow(Rs_grid[0], extent=[Ps[0], Ps[-1], ts[-1], ts[0]], aspect='auto')
```

```python
plt.plot(Ps, np.sum(Rs_grid[0], axis=0))
```

```python
plt.plot(ts, np.sum(Rs_grid[0], axis=1))
```
