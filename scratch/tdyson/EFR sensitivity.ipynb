```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bar charts for Downtime

```python
# operations

groups = ["Objective","Threshold"]
entries = {
    'Initial Ops': (1, 3),
    'Rod Swap': (3, 6),
}

x = np.arange(len(groups)) # the label locations
width = 0.25  # the width of the bars
adjust = 0.25/2
multiplier = 0

fig, ax = plt.subplots()

for attribute, quantity in entries.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, quantity, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Months')
ax.set_xticks(x + width - adjust)
ax.set_xticklabels(groups)
ax.legend(loc='upper left', ncol=3)
ax.set_ylim(0, 7)

plt.show()
```
