```python
import numpy as np
import matplotlib.pyplot as plt

import IPython.display as ipd
```

```python
ts = np.linspace(0,1,1001)

f = 10
N = 1
T = 1/f
ti = 0.2

b = 0*ts
b += np.sin(2*np.pi*f*(ts-ti))
b[:np.argmin(np.abs(ts-ti))] = 0
b[np.argmin(np.abs(ts-ti-N*T)):] = 0
b += np.random.normal(size=ts.shape, scale=0.3)
```

```python
plt.plot(ts,b)
```

```python
ffted = np.fft.rfft(b)
n = b.size
dt = np.diff(ts)[0]
freqs = np.fft.rfftfreq(n, d=dt)
plt.loglog(freqs,np.abs(ffted))
print((freqs[np.argmax(np.abs(ffted))]))
plt.figure()
plt.plot(freqs,np.unwrap(np.angle(ffted)))
```

```python
def Maxwell(x):
    return np.sqrt(2/np.pi) * x**2 * np.exp(-x**2/2)
```

```python
x = np.linspace(0,10,10000)
```

```python
plt.plot(x, Maxwell(x))
```

```python
maxwell = Maxwell(x)
```

```python
ffted = np.fft.rfft(maxwell)
n = maxwell.size
dx = np.diff(x)[0]
freqs = np.fft.rfftfreq(n, d=dx)
plt.loglog(freqs,np.abs(ffted))
```

```python
ipd.Audio(maxwell, rate=44000, autoplay=1)
```

```python
iffted = np.fft.irfft(maxwell)
plt.loglog(np.abs(iffted))
```
